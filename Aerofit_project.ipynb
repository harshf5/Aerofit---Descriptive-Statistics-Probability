#Contional Probability:

Probality of products given gender

```python
print('P(Male) : ' + str(round(d1[0],2)))
print('P(Female) : ' + str(round(d1[1],2)))
print('\n')
males_num=data["Gender"].value_counts()[0]
females_num=data["Gender"].value_counts()[1]

f_281=round(d2['KP281']['Female'] /females_num ,2)
f_481=round(d2['KP481']['Female'] /females_num ,2)
f_781=round(d2['KP781']['Female'] /females_num ,2)

m_281=round(d2['KP281']['Male'] /males_num ,2)
m_481=round(d2['KP481']['Male'] /males_num ,2)
m_781=round(d2['KP781']['Male'] /males_num ,2)


print('P(KP281/Female) : '+  str(f_281))
print('P(KP481/Female) : '+  str(f_481))
print('P(KP781/Female) : '+  str(f_781))
print('\n')
print('P(KP281/Male) : '+  str(m_281))
print('P(KP481/Male) : '+  str(m_481))
print('P(KP781/Male) : '+  str(m_781))
```

Probalities of product given Marital status

```python
d3=data["MaritalStatus"].value_counts(normalize=True)
d3
```

```python
d4=pd.crosstab(index=data['MaritalStatus'],columns=data["Product"],margins=True)
d4
```

```python
print('P(Partnered) : ' + str(round(d3[0],2)))
print('P(Single) : ' + str(round(d3[1],2)))
print('\n')
Partnered_num=data["MaritalStatus"].value_counts()[0]
Single_num=data["MaritalStatus"].value_counts()[1]

p_281=round(d4['KP281']['Partnered'] /Partnered_num ,2)
p_481=round(d4['KP481']['Partnered'] /Partnered_num ,2)
p_781=round(d4['KP781']['Partnered'] /Partnered_num ,2)

s_281=round(d4['KP281']['Single'] /Single_num ,2)
s_481=round(d4['KP481']['Single'] /Single_num ,2)
s_781=round(d4['KP781']['Single'] /Single_num ,2)


print('P(KP281/Partnered) : '+  str(p_281))
print('P(KP481/Partnered) : '+  str(p_481))
print('P(KP781/Partnered) : '+  str(p_781))
print('\n')
print('P(KP281/Single) : '+  str(s_281))
print('P(KP481/Single) : '+  str(s_481))
print('P(KP781/Single) : '+  str(s_781))
```
